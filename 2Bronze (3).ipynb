```python
import os

from pyspark.sql.functions import current_timestamp

basePath="/Volumes/workspace/default/dataflowmed"

rawPath=f"{basePath}/raw"
bronzePath=f"{basePath}/bronze"
```

```python
rawFiles=dbutils.fs.ls(rawPath)
display(rawFiles)
```

```python
bronzeData={}

for file in rawFiles:

    filePath=file.path
    fileName=os.path.basename(filePath)

    if fileName.endswith(".csv"):
        tableName=fileName.replace(".csv","")
        df=spark.read\
            .option("header",True)\
            .option("inferSchema",True)\
            .csv(filePath)
        bronzeData[tableName]=df

    elif fileName.endswith(".json"):
        tableName=fileName.replace(".json","")
        df=spark.read.json(filePath)
        bronzeData[tableName]=df
```

```python
for tableName,df in bronzeData.items():
    print(tableName)
    display(df.limit(5))
```

```python
for tableName,df in bronzeData.items():
    bronzeData[tableName]=df.withColumn("ingestionTime",current_timestamp())
```

```python
for tableName,df in bronzeData.items():
    df.write.format("delta").mode("overwrite").save(f"{bronzePath}/{tableName}")
```
